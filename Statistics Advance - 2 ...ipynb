```python
############################ Statistics Advance - 2 ##############################
#1. What is hypothesis testing in statistics?
#Hypothesis testing is a statistical method used to make decisions based on data.
#It helps us determine whether there's enough evidence in a sample to support a certain belief (hypothesis) about a population.

#Example:
#A company claims their battery lasts 10 hours. You test 50 batteries and find an average life of 9.5 hours.
#You use hypothesis testing to decide:
# Is this difference (0.5 hours) just due to chance, or is the company's claim likely false?

#2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
#Null Hypothesis (H₀): Assumes no effect or no difference. It’s the default assumption.

#Alternative Hypothesis (H₁ or Ha): Represents a new claim, suggesting there is an effect or difference.

#Example:
#H₀: Battery life = 10 hours
#H₁: Battery life ≠ 10 hours (something has changed)
#You test to reject or fail to reject the null hypothesis based on data.

#3. What is the significance level in hypothesis testing, and why is it important?
#Significance Level (α): The threshold for deciding when to reject H₀.

#Common values: 0.05, 0.01, or 0.10.

#It tells you how much risk you’re willing to take in rejecting a true null hypothesis.

#Example:
#If α = 0.05, you accept a 5% risk of concluding there's a difference when there isn't (Type 1 error).

#4. What does a P-value represent in hypothesis testing?
#The P-value tells you the probability of getting a result at least as extreme as the one you observed, if the null hypothesis is true.

#Lower P-value = stronger evidence against H₀.

#5. How do you interpret the P-value in hypothesis testing?
#Compare the P-value to the significance level (α):
#If P-value ≤ α: Reject H₀ (evidence supports Ha)
#If P-value > α: Fail to reject H₀ (not enough evidence)

#Example:
#P-value = 0.03, α = 0.05 → Reject H₀
#P-value = 0.08, α = 0.05 → Fail to reject H₀

#6. What are Type I and Type II errors in hypothesis testing?
#Type I Error (False Positive): Rejecting a true null hypothesis
# Risk = α (significance level)

#Type II Error (False Negative): Failing to reject a false null hypothesis
# Risk = β (depends on sample size and effect size)

#Example:
#Type I: You say the battery is faulty, but it's actually fine.
#Type II: You say the battery is fine, but it’s actually faulty.

#7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
#One-Tailed Test: Tests in one direction only (greater than or less than)
#H₁: μ > μ₀ or μ < μ₀
#Two-Tailed Test: Tests in both directions (difference only)
#H₁: μ ≠ μ₀

#Example:
#One-tailed: “The battery lasts less than 10 hours.”
#Two-tailed: “The battery life is not equal to 10 hours.”

#8. What is the Z-test, and when is it used in hypothesis testing?
#Z-test is used when:
#Population standard deviation (σ) is known
#Sample size is large (n ≥ 30)
#It compares the sample mean to the population mean to check if there's a significant difference.

#9. How do you calculate the Z-score, and what does it represent in hypothesis testing?
Z-score formula:

Z= σ/ n X−μ

 = sample mean
 μ = population mean
 σ = population standard deviation
 n = sample size

#It tells how many standard deviations the sample mean is from the population mean.

#10. What is the T-distribution, and when should it be used instead of the normal distribution?
#Use T-distribution when:

#Sample size is small (n < 30)
#Population standard deviation is unknown
#It is similar to the normal distribution but has fatter tails, which gives more accurate results for small samples.
#11. What is the difference between a Z-test and a T-test?

#Z-test	T-test
#Known σ	Unknown σ
#Large n (≥30)	Small n (<30)
#Uses normal distribution	Uses t-distribution
#12. What is the T-test, and how is it used in hypothesis testing?
#T-test compares means to determine if there's a significant difference.

#Types:
#One-sample t-test: Compare sample mean vs known value
#Two-sample t-test: Compare two sample means
#Paired t-test: Compare before/after for same group

#13. What is the relationship between Z-test and T-test in hypothesis testing?
#Both test whether means differ significantly, but:

#Z-test assumes more info (known σ)
#T-test is more flexible and used more often when σ is unknown

#As sample size increases, t-distribution → normal distribution, so t-test and z-test results become similar.

#14. What is a confidence interval, and how is it used to interpret statistical results?
#A confidence interval (CI) gives a range of values where the true population parameter is likely to fall.
 #For a 95% CI, it means:
#“We are 95% confident the true value lies within this interval.”

#Example:
#Sample mean = 9.5 hours
#95% CI = (9.2, 9.8) → The true battery life is likely between 9.2 and 9.8 hours.
#If the claimed value (e.g., 10) is outside the CI, it suggests a significant difference.

# 15. What is the margin of error, and how does it affect the confidence interval?
# Margin of Error (MoE):
#It represents the range of uncertainty around a sample estimate (like mean or proportion).

# Formula:
#Margin of Error
=𝑍𝛼/2⋅(𝜎𝑛)
Margin of Error=Z α/2​⋅( σ/n)
  
Z 
α/2  = critical value (from Z-table)
σ = population std. dev.
n = sample size

 How It Affects Confidence Interval:
CI=𝑋 ±
Margin of Error
CI= X ±Margin of Error
Larger MoE → wider confidence interval (less precision)
Smaller MoE → narrower interval (more precision)

  Example:
Sample mean = 50, MoE = 3
CI = (47, 53)


# 16. How is Bayes' Theorem used in statistics, and what is its significance?
 # Bayes’ Theorem:
# Used to update probabilities based on new evidence.

#  Formula:
#P(A∣B)= P(B) /P(B∣A)⋅P(A)

#P(A∣B): Probability of A given B
#P(B∣A): Probability of B given A
#P(A): Prior probability of A
#P(B): Total probability of B

 #Significance:
#Bayes' Theorem helps in revising beliefs with incoming data.

# Example:
#A disease test is 99% accurate. What’s the probability someone has the disease if they test positive, given only 1% of the population has it?
#Bayes helps find the real probability, considering false positives.

# 17. What is the Chi-square distribution, and when is it used?
# Chi-square Distribution:
#A distribution used with categorical data to test frequencies or variances.

# Applications:
#Test for independence
#Goodness of fit
#Variance comparison

# Properties:
#Skewed right
#Only positive values
#Depends on degrees of freedom (df)

# 18. What is the Chi-square goodness of fit test, and how is it applied?
#Purpose:
#Tests whether an observed frequency distribution matches an expected distribution.

# Formula:
χ2=∑(O i −E i) 2/Ei
 O 
i
​
 : Observed frequency

𝐸
𝑖
E 
i
​
 : Expected frequency

# Steps:
#Define expected frequencies based on theory
#Collect observed data
#Use the chi-square formula
#Compare with critical value from chi-square table

# Example:
#Roll a die 60 times. Expected frequency = 10 for each side.
#Observed: [8, 12, 9, 11, 10, 10]
#Use χ² test to check if the die is fair.

# 19. What is the F-distribution, and when is it used in hypothesis testing?
# F-distribution:
# A distribution used to compare two variances.

# Used In:
#ANOVA

#F-tests (variance comparison)

# Properties:
#Right-skewed
#Depends on two sets of degrees of freedom
#Only positive values

# 20. What is an ANOVA test, and what are its assumptions?
# ANOVA (Analysis of Variance):
#Used to test differences in means among 3 or more groups.

# Idea:
#Instead of comparing means two at a time (using many t-tests), ANOVA checks all at once.
# Assumptions:
#Independence of observations
#Normal distribution of groups
#Equal variances (homogeneity of variance)

 # Hypotheses:
#H₀: All group means are equal
#H₁: At least one group mean is different
#21. What are the different types of ANOVA tests?

#Type	Purpose
#One-Way ANOVA	Compare means of 3+ independent groups (one factor)
#Two-Way ANOVA	Compare means based on two factors (with or without interaction)
#Repeated Measures	Same group tested multiple times
# 22. What is the F-test, and how does it relate to hypothesis testing?
# F-test:
#Used to compare two variances or test overall variance among groups.
 # Formula:
#𝐹=
# Variance1/
#Variance 2

#F= Variance 2/
#Variance 1
 
#If F is large → variances are significantly different
# Usage in ANOVA:
#ANOVA uses the F-ratio to compare:

#𝐹 =
#Between-group variance
#Within-group variance
#F= 
#Within-group variance
#Between-group variance

#If the ratio is high → at least one group mean is likely different.

#Summary Chart:

#Concept	Purpose
#Margin of Error	Range of uncertainty in estimates
#Bayes' Theorem	Update probabilities with new data
#Chi-square	Test for independence/goodness of fit
#F-distribution	Compare variances
#ANOVA	Compare multiple group means
#F-test	Tests overall variance or mean differences

 
```
